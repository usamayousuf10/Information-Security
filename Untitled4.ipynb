```python
import time
from itertools import product
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives import padding
from random import randint
import pandas as pd
import threading

def encrypt(key, iv, text):
    backend = default_backend()
    cipher = Cipher(algorithms.TripleDES(key), modes.CBC(iv), backend=backend)
    encryptor = cipher.encryptor()
    return encryptor.update(text) + encryptor.finalize()

def decrypt(key, iv, ciphertext):
    backend = default_backend()
    cipher = Cipher(algorithms.TripleDES(key), modes.CBC(iv), backend=backend)
    decryptor = cipher.decryptor()
    return decryptor.update(ciphertext) + decryptor.finalize()



def main():
    start_time = time.time()
    text = "My name is Usama Yousuf Khan I am currently studying Computer Science in Fast University Please dont tell anyone"
    iv = bytearray(8)
    for i in range(8):
        iv[i] = randint(0, 255)

    key1_list = list(product(range(2), repeat=56))
    key2_list = list(product(range(2), repeat=56))

    df = pd.DataFrame(columns=['key1', 'key2', 'encrypt', 'decrypt'])
    i = 0
    for key1 in key1_list:
        key1_bytearray = bytearray(8)
        for j in range(8):
            for k in range(7):
                key1_bytearray[j] += key1[(j * 7) + k] * (2 ** (6 - k))
        for key2 in key2_list:
            key2_bytearray = bytearray(8)
            for j in range(8):
                for k in range(7):
                    key2_bytearray[j] += key2[(j * 7) + k] * (2 ** (6 - k))
            ciphertext = encrypt(key1_bytearray, iv, text)
            plaintext = decrypt(key2_bytearray, iv, ciphertext)
            if plaintext == text:
                df.loc[i] = [key1_bytearray,key2_bytearray,ciphertext,plaintext]
                i += 1
                break

    print("--- %s seconds ---" % (time.time() - start_time))
    print(df)
    
    
```
